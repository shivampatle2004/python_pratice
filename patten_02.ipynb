```python
n=int(input("e a n"))
for i in range(1,n+1):
  for j in range(1,i+1):
    print(j,end=" ")
  for k in range(j-1,0,-1):
      print(k,end=" ")

  print("\n")
```

```python
n=int(input("enter the length of palindro"))
for i in range(1,n+1):
  for j in range(n-i):
    print(" ",end=" ")
  for k in range(i,0,-1):
      print(k,end=" ")

  for l in range(2,i+1):
    print(l,end=" ")
  print("\n")
```
