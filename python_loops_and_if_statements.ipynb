Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below: Salary(Lakhs) : Tax(%)

Below 5 : 0% 5-10 : 10% 10-20 : 20% aboove 20 : 30%

```python
n = int(input("Enter your annual salary :\n"))

har = 0.1 * n
da = 0.05 * n
pf = 0.03 * n
ist = 0
flag = 0

if 500000 <= n < 1000000:
    ist = 0.1 * n
elif 1000000 <= n < 2000000:
    ist = 0.2 * n
elif n >= 2000000:
    ist = 0.3 * n
else:
    print("Note: Salary below 500,000 - No income tax applied")
    flag = 1

if flag == 0:
    deduct = har + da + pf + ist
    total = n - deduct
    print("In Hand Salary Is :", total)
else:
    print("In hand salary is:", n)
```

Write a program that take a user input of three angles and will find out whether it can form a triangle or not.

Note: The Fibonacci Sequence is a series of numbers. The next number is found by adding up the two numbers before it. The first two numbers are 0 and 1. For example, 0, 1, 1, 2, 3, 5, 8, 13, 21. The next number in this series above is 13+21 = 34

```python
n=int(input("enter the number till you want fibonacci seris "))
print(0)
print(1)
a,b,i=0,1,1
for i in range(n-2):
  c=a+b
  print(c)
  a=b
  b=c

```

Find the factorial of a given number.
Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: !) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```python
n=int(input("eneter the number for till you wantt factorical:"))
s=1

for ij in range(1,n+1):
  s=s*ij

print(s)
```
