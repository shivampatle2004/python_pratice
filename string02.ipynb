Create Short Form from initial character

```python
s=input()
r=''
r=r+s[0]
for i in range(1,len(s)):
  if(s[i]==' '):
    r=r+s[i+1]


print(r.upper()  )
```

Append second string in the middle of first string

```python
s=input()
t=input()
c=0
r=''
n1=len(s)//2
n2=abs(len(s)-n1)
for i in range(0,n1):
  r=r+s[i]
for j in range(n1,n1+len(t)):
  r=r+t[c]
  c=c+1
for k in range(len(t)+n1,len(t)+n1+n2)  :
  i=i+1
  r=r+s[i]
print(r)
```
