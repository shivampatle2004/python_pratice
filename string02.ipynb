Create Short Form from initial character

```python
s=input()
r=''
r=r+s[0]
for i in range(1,len(s)):
  if(s[i]==' '):
    r=r+s[i+1]


print(r.upper()  )
```

```python
s=input()
t=input()
c=0
r=''
n1=len(s)//2
n2=abs(len(s)-n1)
for i in range(0,n1):
  r=r+s[i]
for j in range(n1,n1+len(t)):
  r=r+t[c]
  c=c+1
for k in range(len(t)+n1,len(t)+n1+n2)  :
  i=i+1
  r=r+s[i]
print(r)
```

Given string contains a combination of the lower and upper case letters. Write a program to arrange the characters of a string so that all lowercase letters should come first.

```python
s=input()
r=''
r1=''
for i in s:
  if(i.upper()==i):
    r=r+i
  else:
    r1=r1+i
print(r1+r)




```

Take a alphanumeric string input and print the sum and average of the digits that appear in the string, ignoring all other characters.

```python
s=input()
r=''
for i in s:
 if(i.isdigit()):
  r=r+i
s=0

for j in r:
  s=s+int(j)
av=s/len(r)
print(av)
print(s)
```
